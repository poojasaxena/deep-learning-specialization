## 4 - Neural Network model

Logistic regression did not work well on the "flower dataset". You are going to train a Neural Network with a single hidden layer.

**Here is our model**:
<img src="images/classification_kiank.png" style="width:600px;height:300px;">

**Mathematically**:

For one example $x^{(i)}$:
$$z^{[1] (i)} =  W^{[1]} x^{(i)} + b^{[1]}\tag{1}$$ 
$$a^{[1] (i)} = \tanh(z^{[1] (i)})\tag{2}$$
$$z^{[2] (i)} = W^{[2]} a^{[1] (i)} + b^{[2]}\tag{3}$$
$$\hat{y}^{(i)} = a^{[2] (i)} = \sigma(z^{ [2] (i)})\tag{4}$$
$$y^{(i)}_{prediction} = \begin{cases} 1 & \mbox{if } a^{[2](i)} > 0.5 \\ 0 & \mbox{otherwise } \end{cases}\tag{5}$$

Given the predictions on all the examples, you can also compute the cost $J$ as follows: 
$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large\left(\small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right)  \large  \right) \small \tag{6}$$

**Reminder**: The general methodology to build a Neural Network is to:
    1. Define the neural network structure ( # of input units,  # of hidden units, etc). 
    2. Initialize the model's parameters
    3. Loop:
        - Implement forward propagation
        - Compute loss
        - Implement backward propagation to get the gradients
        - Update parameters (gradient descent)

You often build helper functions to compute steps 1-3 and then merge them into one function we call `nn_model()`. Once you've built `nn_model()` and learnt the right parameters, you can make predictions on new data.

```python
### defining the Neural Network structure
# n_x : sixe of input layer
# n_h : size of hidden layer
# n_y : size of output layer

def layer_sizes(X, Y):
    n_x = X.shape[0]
    n_h = 4
    n_y = Y.shape[0]
    
    return (n_x, n_h, n_y)
```

```python
def initialize_parameters(n_x, n_h, n_y):
    np.random.seed(2)
    
    ### layer-1/Hidden Layer
    W1 = np.random.randn(n_h,n_x)
    b1 = np.zeros((n_h, 1))

    ### layer-2/ Output layer
    W2 = np.random.randn(n_y, n_h)
    b2 = np.zeros((n_y, 1))
    
    
   # assert(W1.shape == (n_h, n_x))
   # assert(W2.shape == (n_y, n_h))
   # assert(b1.shape == (n_h, 1))
   # assert(b2.shape == (n_y, 1))
    
    
    parameters = {
        "W1": W1,
        "b1": b1,
        "W2": W2,
        "b2": b2
    }
    
    return parameters
```

```python
def sigmoid(z):
    return 1/ (1+ np.exp(-z))
```

```python
def get_diff_tanh(z):
    return 1 - np.tanh(z)^2
```

```python
def forward_propagation(X, Y, parameters):
    
    
    Z1 = np.dot(parameters["W1"], X) + parameters["b1"]
    A1 = np.tanh(Z1)
    
    Z2 = np.dot(parameters["W2"], A1) + parameters["b2"]
    A2 = sigmoid(Z2)
    
    #print("A1.shape :"+ str(A1.shape)+" A2.shape:"+str(A2.shape))
    #assert(A1.shape == A2.shape)
    
    cache= {
        "Z1":Z1,
        "W1":parameters["W1"],
        "b1":parameters["b1"],
        "Z2":Z2,
        "W2":parameters["W2"],
        "b2":parameters["b2"]
    }
    
    
    return cache, A1, A2
```

```python
def backward_propagation(X,Y, parameters):
    m = X.shape[1]

        
    ## output, Z1, W1, b1, Z2, W2, b2, 
    cache, A1, A2 = forward_propagation(X, Y, parameters)
    
    cost = -1/m * (np.dot(Y,np.log(A2).T) + np.dot(1-Y, np.log(1-A2).T) )
    
    ### for layer-2
    dZ2 = A2 - Y
    dW2 = 1/m * np.dot(dZ2, A1.T)
    db2 = 1/m * np.sum(dZ2)
    
    
    ### for layer-1
    dZ1 = np.dot(W2, dZ1.T) * get_diff_tanh(Z1)
    dW1 = 1/m * np.dot(dZ1, X.T)
    db1 = 1/m * dZ1
    
    dParam = {
        "dW1": dW1,
        "db1": db1,
        "dW2": dW2,
        "db2": db2
    }
    
    return cost, dParam
```

```python
def optimize(X, learning_rate, no_iteration):
    
    ## get the Neural Network Size
    n_x = layer_sizes(X, Y)[0]
    n_h = layer_sizes(X, Y)[1]
    n_y = layer_sizes(X, Y)[2]
    
    print(isinstance(n_x, int))
    print(isinstance(n_h, int))
    print(isinstance(n_y, int))

    print("n_x: "+str(n_x)+" n_h :"+str(n_h)+" n_y:"+str(n_y))
    ## initialize Parameters
    parameters = initialize_parameters(int(n_x), int(n_h), int(n_y))
    
    costs= []
    
    for i in range(no_iteration):
        
        ## forward/propagation
        cache, A1, A2 = forward_propagation(X, Y, parameters)
    
        ## backward/propagation 
        cost, dParam = backward_propagation(X,Y, parameters)
        
        ## Now, find the difference of the dW's, db's
        W1 = W1 - learning_rate * dParam["dW1"]
        b1 = b2 - learning_rate * dParam["db1"]
        
        W2 = W2 - learning_rate * dParam["dW2"]
        b2 = b2 - learning_rate * dParam["db2"]
        
        if(i % 100 == 0):
            costs.append[cost]
        
        
    return costs    
```
