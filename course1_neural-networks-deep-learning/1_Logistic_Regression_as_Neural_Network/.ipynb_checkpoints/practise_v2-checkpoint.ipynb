```python
## lets change the train/test, lets flatten it
train_x_flatten = train_set_x_orig.reshape(train_set_x_orig.shape[0], -1).T
test_x_flatten = test_set_x_orig.reshape(test_set_x_orig.shape[0], -1).T

print("train_flatten shape : " + str(train_x_flatten.shape))
print("test_flatten shape : " + str(test_x_flatten.shape))
```

```python
## lets normalize, in this case, divide by 255 
train_x = train_x_flatten/255.
test_x = test_x_flatten/255.
train_x[:, 1]
```

```python
## define sigmoid
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
## initialize the W, b
def initialize(dim):
    ## W has the dimension (dim, 0)
    ## b has the value 0
    
    W = np.zeros((dim, 1))
    b = 0
    
    assert(W.shape == (dim,1))
    assert(isinstance(b, float) or isinstance(b, int))
    
    return W,b
```

```python
### forward propagation

def forward_propagation(X, Y, W, b, no_of_sample):    
    Z = np.dot(W.T, X) + b
    A = sigmoid(Z)
    
    J = -1/no_of_sample * (np.dot(Y, np.log(A).T) + np.dot(1-Y, np.log(1-A).T))
    
    cache = {
        "A" : A,
        "cost" : J
    }
                    
    return cache
```

```python
def backward_propagation(X, Y, A, no_of_sample):
    
    dZ = A - Y
    assert(dZ.shape == A.shape)
    
    dW = 1/no_of_sample * np.dot(X, dZ.T)
    db = 1/no_of_sample * np.sum(dZ)
    
    assert(dW.shape == (X.shape[0],1))
    assert(isinstance(db, float))
    
    dParam = {
        "dW" : dW,
        "db": db
    }
    
    return dParam
```

```python
def optimize(X, Y, no_of_iteration, learning_rate, dim, no_of_samples, debug= False):    

    ## empty array to hold cost value
    costs = []
    
    ## initialize W, b to zeros
    W, b = initialize(dim)
    
    if(debug):
        print("before loop W  and b is :" + str(W[1:3,:].tolist()) + " "+ str(b))        
        
    ## lets iterate, basically converge the gradiant descent
    for i in range(no_of_iteration):    

        ## forward propagation, input: W, b, m; Output: A, cost
        cache = forward_propagation(X, Y, W, b, no_of_samples) 
        A = cache["A"]
        cost = cache["cost"]

        if(debug):
            print("======================== i:"+ str(i))
            print("W and b : " +str(W[1:3,:].tolist()) + str(b))
            print("cost : " + str(cost), " A :" + str(A[:, 1:4].tolist()))

        ## backward propagation, inputs: X, Y, A, m; Outputs: dW, db
        dparams = backward_propagation(X, Y, A, no_of_samples)
        dW = dparams["dW"]
        db = dparams["db"]    

        if(debug):
            print("dW :" + str(dW[1:3,:].tolist()) +" db : " + str(db))
        
        if i % 100 == 0:
            costs.append(cost)
        
        ## lets calculate the new W, b 
        W =  W - learning_rate * dW
        b =  b - learning_rate * db

        assert(dW.shape == W.shape)
        assert(isinstance(b, float))
    
    param = {
        "W" : W,
        "b" : b
          }
    grads = {
        "dW": dW,
        "db": db
    }
        
    return param, grads, costs
```

```python
def prediction(W, b, X, debug=False):
    
    ### final A, input: X,W,b; Output: A
    A = sigmoid(np.dot(W.T, X) + b)
    
    ## lets calculate the prediction
    prediction = np.zeros((1, X.shape[1]))
    
    for i in range (A.shape[1]):
        if(debug):
            print(" A["+str(i)+"] :" + str(A[:,i]))
        prediction[:,i] = np.where(A[:,i]> 0.5, 1, 0)
        
    ## lets calculate the accuracy
    assert(prediction.shape == A.shape)
    
    return prediction
```

```python
def get_accuracy(Y_predicted, Y_real):
    ## lets calculate the accuracy
    assert(Y_predicted.shape == Y_real.shape)
    
    accuracy = (100 - np.sum(np.abs(Y_predicted - Y_real))/Y_real.shape[1] * 100)
    assert(isinstance(accuracy, float))
    
    return accuracy
```

```python
def model_nn(X_train, Y_train, X_test, Y_test, no_of_iteration, learning_rate, debugSession=True):
    
    dim = X_train.shape[0]
    m = X_train.shape[1]
    
    if(debugSession):
        print("dim :"+str(dim)+ " no_of_sample :"+str(m))
    param, grads, costs = optimize(X_train, Y_train, no_of_iteration, learning_rate, dim, no_of_samples=m, debug= False)

    Y_prediction_train = prediction(param["W"], param["b"], X_train, debug=False)
    Y_prediction_test = prediction(param["W"], param["b"], X_test, debug=False)
    
    accuracy_train =  get_accuracy(Y_prediction_train, Y_train)
    accuracy_test = get_accuracy(Y_prediction_test, Y_test)
                
    print("accuracy on train dataset is :", accuracy_train)
    print("accuracy on test dataset is : ",accuracy_test)
    
    d = {
         "costs":costs,
         "Y_prediction_train":Y_prediction_train,
         "Y_prediction_test":Y_prediction_test,
        "W":param["W"],
        "b":param["b"],
        "learning_rate":learning_rate,
        "no_of_iteration":no_of_iteration        
    }

    return d
```

```python
d = model_nn(train_x, train_set_y, test_x, test_set_y, no_of_iteration=2000, learning_rate=0.005, debugSession=False)
```

```python
costs = np.squeeze(d["costs"])
plt.plot(costs)
plt.ylabel('cost')
plt.xlabel('iterations (per hundreds)')
#plt.title("Learning rate ="  "learning_rate"))
plt.show()
```

```python
## example of a picture that was wrongly classified
index = 5
plt.imshow(test_x[:, index].reshape((num_px, num_px, 3)))
print("y =" + str(test_set_y[:,index]) + " and you predicted :"+classes[int(np.squeeze(d["Y_prediction_test"][0,index]))].decode("utf-8"))
```

```python
learning_rates = [ 0.01, 0.005, 0.001]
models = {}
for i in learning_rates:
    print("learning rate is :" + str(i))
    models[i]= model_nn(train_x, train_set_y, test_x, test_set_y, no_of_iteration=2000, learning_rate=i, debugSession=False)
    print ('\n' + "-------------------------------------------------------" + '\n')

for i in learning_rates:
    plt.plot(np.squeeze(models[i]["costs"]), label=models[i]["learning_rate"])
plt.ylabel('cost')
plt.xlabel('iterations (hundreds)')
legend = plt.legend(loc='upper center', shadow=True)
frame = legend.get_frame()
frame.set_facecolor('0.90')
plt.show()
```
