```python
def sigmoid(z) :
    return 1/ ( 1 + np.exp(-z))
```

```python
## initilize the variables , W, b
def initialize_parameters(dim) :
    W = np.zeros((dim,1))
    b = 0
    
    assert (W.shape == (dim,1))
    assert (isinstance(b, float) or isinstance(b, int))
    
    return W,b
```

```python
def propagate(X, Y, W, b, debug=False):
    
    m = X.shape[1]
    if(debug):
        print("m  is %d " %m)
    
    ## forward propagation from X to J 
    Z = np.dot(W.T, X) + b ## (12288,1).T x(12288,129) ~ (1, 129)
    A = sigmoid(Z)
    if (debug):  
        print("A shape :" + str(A.shape))
    cost = -1/m * (np.dot(Y, np.log(A).T) + np.dot((1-Y), np.log(1-A).T))
    if(debug):
        print("cost shape :" + str(cost.shape))
    
    ## back propagation from J to W, b
    dW = 1/m * np.dot(X, (A-Y).T) ## (12288, 129) x (1, 129).T ~ (12288,1)
    db = 1/m * np.sum(A-Y)   ## (1, 129)
    
    if(debug):
        print("W shape is :"+ str(W.shape) +" and dW shape is :" + str(dW.shape))
    
    assert(dW.shape == W.shape)
    assert(db.dtype == float)
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    
    grads = {
        "dW":dW,
        "db":db        
    }
    
    return grads, cost
```

```python
## we habe calculated the dW, db and the cost for single iteration, 
## we need to iterate till the W,b converge, lets iterate
def optimize(W, b, X, Y, num_iteration, learning_rate, print_cost = False):
    
    costs = []
    
    for i in range(1, num_iteration):
            grads, cost = propagate(X, Y, W, b)
            dW = grads["dW"]
            db = grads["db"]
            
            W = W - learning_rate * dW
            b = b - learning_rate * db

            ## record the cost
            costs.append(cost)
            
            if print_cost:
                print("cost after iteration %i : %f" %(i, cost))
                
    params = {
        "W": W, 
        "b": b
    }
    
    grads= {
        "dW":dW,
        "db":db
    }
    
    return params, grads, costs
```

```python
def predict(W,b,X):    

    A = sigmoid(np.dot(W.T,X) + b)
    
    m = X.shape[1]
    Y_prediction = np.zeros((1, m))
    
    for i in range(A.shape[1]):
        Y_prediction[0,i] = np.where(A[0,i]> 0.5, 1, 0)
        
    assert(Y_prediction.shape == (1,m))
    
    return Y_prediction
```

```python
def model_nn(X_train, Y_train, num_iteration, learning_rate, print_cost):
    
    
    ## initialize the parameters
    dim = X_train.shape[0]
    W, b = initialize_parameters(dim)
    print("%d is the dimension" %(dim))
    
    grads, cost = propagate(X_train, Y_train, W, b, debug=True)
    W = grads["dW"]
    b = grads["db"]
    
    params, grads, cost = optimize(W, b, X_train, Y_train, num_iteration, learning_rate, print_cost)
    
    Y_prediction = predict(params["W"], params["b"],X_train)
    
    return Y_prediction
    
```
